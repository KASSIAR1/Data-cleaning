# Supprimer les doublons

```python
#Suppression des valeurs dupliquées
df.drop_duplicates()
```

```python
#Supprimer et enregistrer cette tache dans le tableau d'entré
df = df.drop_duplicates()
df
```

```python
#Supprission de la colonne inutile et enregistrement de la tache dans la table d'entrée 
df = df.drop(columns = "Not_Useful_Column")
df
```

 # Nettoyage de la colonne [ Last_Name ]

```python
df["Last_Name"]
```

```python
# Supprimer les slash (ligne 2)
df["Last_Name"] = df["Last_Name"].str.lstrip('/')
df
```

```python
# Supprimer les points (ligne 11)
df["Last_Name"] = df["Last_Name"].str.lstrip("...")
df["Last_Name"]
```

```python
# Supprimer les tirets (ligne 14)
df["Last_Name"] = df["Last_Name"].str.rstrip("_")
df["Last_Name"]
```

```python
# Check de la table après ces traitements
df
```

```python
# Combiner les traitements précedents en une seule ligne
```

```python
df['Last_Name'] = df['Last_Name'].str.strip('123._/') # il faudra annuler les précédents traitement de compiler cette ligne;
```

 # Nettoyage de la colonne [ Phone_Number ]

```python
#Supprimer les valeurs non numériques dans la colonne
df["Phone_Number"] = df["Phone_Number"].str.replace('[^0-9]', '')
```

```python
df["Phone_Number"]
```

```python
# Mettre en forme la colonne 
```

```python
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x))
```

```python
df["Phone_Number"]
```

```python
# Remplacer les valeurs nulle (nan) 
df["Phone_Number"] = df["Phone_Number"].str.replace('nan----','')
df["Phone_Number"] = df["Phone_Number"].str.replace('NAN----','')
df["Phone_Number"] = df["Phone_Number"].str.replace('--','')
df["Phone_Number"] = df["Phone_Number"].str.replace('-','')
df
```

# Décomposition de la colonne [ Adress ]

```python
#Decomposition de la colonne adresse en different 3 colonnes ()
df[["Street_Address", "State", "Zip_Code"]] = df["Address"].str.split(',',2, expand=True)
df
```

# Nettoyage de la colonne [Do_Not_Contact]

```python
#Remplacer les valeurs
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('Yes','Y')

df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('No','N')
df
```

```python
# Supprimer les nan
df=df.fillna('')
df
```

# Supprimer les contacts qui n'ont pas de numéro de téléphone dans la table

```python
for x in df.index:
    if df.loc[x, "Phone_Number"] == '':
        df.drop(x, inplace=True)

df
```

# Supprimer les contacts qui n'ont pas été contactés

```python
for x in df.index:
    if df.loc[x, "Do_Not_Contact"] == 'Y':
        df.drop(x, inplace=True)

df
```

# Reset_Index

```python
df = df.reset_index(drop=True)
df
```
